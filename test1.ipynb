```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
from skimage.morphology import watershed
from skimage.feature import peak_local_max
from sklearn.cluster import MeanShift


img = cv2.imread('./data/DIC-C2DH-HeLa/Sequence 1/t002.tif',cv2.IMREAD_UNCHANGED)

plt.figure('test1')
plt.imshow(img)
plt.axis('off')
plt.title('test1')
plt.show()

hist,bins = np.histogram(img.flatten(),256,[0,256])
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.show()
```

```python

sobelX = cv2.Sobel(img,cv2.CV_64F,1,0)#x
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1)#y

sobelX = np.uint8(np.absolute(sobelX))#x
sobelY = np.uint8(np.absolute(sobelY))#y

sobelCombined = cv2.bitwise_or(sobelX,sobelY)#
plt.figure('sob')
plt.imshow(sobelCombined,cmap='gray')
plt.axis('off')
plt.title('sob')
plt.show()
```

```python
hist,bins = np.histogram(sobelCombined.flatten(),256,[0,256])
plt.hist(sobelCombined.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.show()
```

```python
res,output = cv2.threshold(sobelCombined,60,255,cv2.THRESH_BINARY)
plt.figure('output')
plt.imshow(output,cmap='gray')
plt.axis('off')
plt.title('output')
plt.show()



```
