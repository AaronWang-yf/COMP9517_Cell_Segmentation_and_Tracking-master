```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.segmentation import watershed
from skimage.feature import peak_local_max
from scipy import ndimage as ndi
```

```python
# based on Raymond‘s result
img_path = "SEG/seq1/seg_Sequence 1_t006.tif"
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
img1 = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img,'gray')
```

```python
ret, img_threh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
blurred = cv2.GaussianBlur(img_threh, (9,9),0)
plt.imshow(blurred,'gray')
                            
```

```python
kernel=np.ones((5,5),np.uint8) 
erosion=cv2.erode(blurred,kernel,iterations=1) 
dilation=cv2.dilate(erosion,kernel,iterations=1) 
ret, thresh = cv2.threshold(dilation, 190, 255, cv2.THRESH_BINARY)
plt.imshow(thresh,'gray')
```

```python
distance = ndi.distance_transform_edt(thresh)
local_maxi = peak_local_max(distance, indices=False, footprint=np.ones((55,55)),
                            labels=thresh)
markers = ndi.label(local_maxi)[0]
labels = watershed(-distance, markers, mask=thresh)
plt.imshow(labels, cmap=plt.cm.nipy_spectral)
```
