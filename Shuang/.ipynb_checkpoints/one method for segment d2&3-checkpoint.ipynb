```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from libtiff import TIFF
```

```python
def min_filter(image_I, N):
    (height, width) = image_I.shape
    image_min_filter = np.zeros((height, width), dtype=np.uint8)
    init_neighbourhood = np.zeros((N,N))
    padding = N // 2
    init_neighbourhood[padding, padding] = 0
    imageI_p = cv2.copyMakeBorder(image_I, padding, padding, padding, padding, cv2.BORDER_CONSTANT, dst=None, value=255)
 
    for h in range(padding, (padding+height)):
        for w in range(padding, (padding+width)):
            up = h - padding
            down = h + padding + 1
            left = w - padding
            right = w + padding + 1
            update_neighbourhood = imageI_p[up:down, left:right] + init_neighbourhood
            image_min_filter[h - padding][w - padding] = np.min(update_neighbourhood)
#     plt.imshow(image_min_filter, cmap=plt.cm.gray)
    return image_min_filter
```

```python
def max_filter(image_I, N):
    (height, width) = image_I.shape
    image_max_filter = np.zeros((height, width), dtype=np.uint8)
    init_neighbourhood = np.zeros((N,N))
    padding = N // 2
    init_neighbourhood[padding, padding] = 0
    imageI_p = cv2.copyMakeBorder(image_I, padding, padding, padding, padding, cv2.BORDER_CONSTANT, dst=None, value=0)

    for h in range(padding, (padding+height)):
        for w in range(padding, (padding+width)):
            up = h - padding
            down = h + padding + 1
            left = w - padding
            right = w + padding + 1
            update_neighbourhood = imageI_p[up:down, left:right] + init_neighbourhood
            image_max_filter[h - padding][w - padding] = np.max(update_neighbourhood)
#     plt.imshow(image_max_filter, cmap=plt.cm.gray)
    return image_max_filter
```

```python
def subtraction(image_I,image_B):
    (height, width) = image_I.shape
    image_O = np.zeros((height, width), dtype=np.float32)
    image_I_float = image_I.astype('float32')
    image_B_float = image_B.astype('float32')
    for h in range(height):
        for w in range(width):
            image_O[h][w] =  image_I_float[h][w] - image_B_float[h][w]
#             image_O[h][w] = cv.subtract(image_I_float[h][w], image_B_float[h][w])
    a = 0
    b = 255
    c = np.min(image_I)
    d = np.max(image_I)
    image_O = (image_O - c) * ((b - a) / (d - c)) + a
#     plt.imshow(image_O, cmap=plt.cm.gray)
    return image_O
```

```python
# for dataset 2 and 3
image = cv2.imread("t000.tif",0)
plt.imshow(image,'gray')
```

```python

N = 85
image_min_filter = min_filter(image, N)
image_max_filter = max_filter(image_min_filter,N)
image_O = subtraction(image,image_max_filter)
```

```python
IMG_OUT = cv2.cvtColor(image_O.astype('uint8'), cv2.COLOR_GRAY2RGB)
plt.imshow(IMG_OUT,'gray')
```

```python
gray_image = cv2.cvtColor(IMG_OUT,cv2.COLOR_BGR2GRAY) 
plt.imshow(gray_image,'gray')
```

```python
blurred = cv2.GaussianBlur(gray_image, (7,7),0)
plt.imshow(blurred,'gray')
```

```python
(_, thresh) = cv2.threshold(blurred,65, 255, cv2.THRESH_BINARY)
plt.imshow(thresh,'gray')
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
closed = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
plt.imshow(closed,'gray')
```

```python
# closed = cv2.erode(closed, None, iterations=3)
# closed = cv2.dilate(closed, None, iterations=3)

# blurred = cv2.GaussianBlur(closed, (3, 3),0)
# (_, thresh) = cv2.threshold(blurred,120, 255, cv2.THRESH_BINARY)
# plt.imshow(thresh,'gray')
```

```python
(_, cnts, _) = cv2.findContours(closed.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
c = sorted(cnts, key=cv2.contourArea, reverse=True)
draw_img = IMG_OUT.copy()
for i in range(len(c)):
    rect = cv2.minAreaRect(c[i])
    box = np.int0(cv2.boxPoints(rect))
    draw_img = cv2.drawContours(draw_img, [box], -1, (0, 0, 255),2)

plt.imshow(draw_img,'gray')
```
